```python
# Para estÃ¡tico
#%matplotlib inline
# Para dinÃ¡mico
%matplotlib notebook 
import numpy as np

import matplotlib.pyplot as plot

time        = np.arange(0, 10, 0.1);
amplitude   = np.sin(time)

plot.plot(time, amplitude)
plot.title('FunciÃ³n de seno')
plot.xlabel('Tiempo')
plot.ylabel('Amplitud')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')

plot.show()
```
